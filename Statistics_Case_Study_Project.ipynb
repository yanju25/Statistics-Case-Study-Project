```python
plt.figure(figsize = (10,5))
sns.scatterplot(data = df,x = 'age', y = 'salary', color = 'blue',edgecolor = 'linen',alpha = 0.5 )
plt.title("Scatterplot of Employee Age v/s Employee Salary")
plt.xlabel('Employee Age')
plt.ylabel('Employee Salary')
plt.show()
```

```python
plt.figure(figsize = (10,5))
sns.scatterplot(data = df,x = 'healthy_eating', y = 'active_lifestyle', color = 'Red',edgecolor = 'linen',alpha = 0.5 )
plt.title("Scatterplot of healthy_eating v/s active_lifestyle")
plt.xlabel('healthy_eating')
plt.ylabel('active_lifestyle')
plt.show()
```

```python
plt.figure(figsize = (10,5))
sns.scatterplot(data = df,x = 'healthy_eating', y = 'salary', color = 'k',edgecolor = 'linen',alpha = 0.5 )
plt.title("scatterplot of healthy_eating v/s Employee Salary")
plt.xlabel('healthy_eating')
plt.ylabel('Employee Salary')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data = df, x = 'groups',edgecolor = 'linen',alpha = 0.7)
plt.title("count plot of groups")
plt.xlabel('Groups')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize = (11,6))
sns.histplot(data = df, x = 'salary',color = 'orange', edgecolor = 'linen',alpha = 0.7, bins = 10)
plt.title("histogram of salary")
plt.xlabel('salary')
plt.ylabel('count')
plt.show()
```
